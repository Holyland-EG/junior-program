35. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –º–µ—Ç–æ–¥ Interquartile Range (IQR) (‚òÖ‚òÖ‚òÖ)

–ü–æ—á–∏—Ç–∞—Ç—å –ø—Ä–æ –Ω–µ–≥–æ –º–æ–∂–Ω–æ –ø–æ —Å—Å—ã–ª–∫–µ –≤—ã—à–µ.

```python
s = pd.Series(np.logspace(-2, 2, 30))

print(s)

kek = s.mean()
che = s.std()

FirstQ = kek - .6745*che 
ThirdQ = kek + .6745*che 

IQR = ThirdQ - FirstQ
print('IQR = ', IQR, '\n')

a = FirstQ - 1.5*IQR
b = ThirdQ + 1.5*IQR

print('Boxplot lower whisker: ',a)
print('Boxplot upper whisker: ', b)


for i in range(len(s)):
    if s[i] < a:
        print('Outlier = ', s[i])
    elif s[i] > b:
        print('Outlier = ', s[i])
    else:
        print('Norm')
```

* –í—ã–≤–µ—Å—Ç–∏ —É –∫–∞–∫–æ–π –º–æ–¥–µ–ª–∏, –∫–∞–∫–æ–≥–æ –∏–∑–≥–æ—Ç–æ–≤–∏—Ç–µ–ª—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–∞—è –≤–º–µ—Å—Ç–∏–º–æ—Å—Ç—å –ø–æ –ø–∞—Å—Å–∞–∂–∏—Ä–∞–º

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

#1 –ó–∞–º–µ–Ω—è–µ–º –≤—Å–µ —Ç–æ—á–∫–∏ –≤ –Ω–∞–∑–≤–∞–Ω–∏–∏ –Ω–∞ '_'

df.columns = [col.replace('.', '_') for col in df.columns]

df
```

```python
#2 –í—ã–¥–µ–ª–∏—Ç—å –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏

text_columns =  df.select_dtypes(include = ['object']) 
numeric_columns =  df.select_dtypes(exclude = ['object']) 

print(text_columns.describe())
print()
print(numeric_columns.describe())
```

```python
#3 –ü–æ—Å—á–∏—Ç–∞—Ç—å –∫–æ–ª-–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤–æ –≤—Å–µ—Ö —Ç–µ–∫—Å—Ç–æ–≤—ã—Ö\–∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö.

for i in numeric_columns:
    print(i, df[i].nunique())
print()
for i in text_columns:
    print(i, df[i].nunique())
```

```python
#4 –ö–æ–ª–æ–Ω–∫–∏ —Å —Ç–æ–ª—å–∫–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ —É–¥–∞–ª—è–µ–º (–Ω–∞–ø—Ä–∏–º–µ—Ä Model –∏ Make)
for i in text_columns:
    if df[i].count() == df[i].nunique():
        df = df.drop([i], axis = 1)
df
```

```python
for i in numeric_columns:
    if df[i].count() == df[i].nunique():
        df = df.drop([i], axis = 1)
#df
```

```python
#5 –î–ª—è —Å—Ç–æ–ª–±—Ü–æ–≤ —É –∫–æ—Ç–æ—Ä—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –±–æ–ª–µ–µ 10, –æ—Å—Ç–∞–≤–ª—è–µ–º —Ç–æ–ø 10 –∑–Ω–∞—á–µ–Ω–∏–π, –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∑–∞–º–µ–Ω—è–µ–º –Ω–∞ 'other'
for i in df.columns:
    val = df[i].nunique()
    if val > 10:
        if df[i].count() > 10:
            df = df.replace(df[i].value_counts().index[10:], 'other')
        else:
            continue
    else:
        continue
pd.set_option('max_rows',100)
print(df)
```

```python
#6 –ß–∏—Å–ª–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã –ø—Ä–æ–≤–µ—Ä—è–µ–º –Ω–∞ –ø—Ä–æ–ø—É—Å–∫–∏. –ü–æ–¥—Å—á–∏—Ç—ã–≤–∞–µ–º –∫–æ–ª-–≤–æ nan\none –∏ –¥–µ–ª–∏–º –Ω–∞ —Ä–∞–∑–º–µ—Ä –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞ 
#-—Å–æ—Ö—Ä–∞–Ω—è–µ–º –≤ –æ—Ç–¥–µ–ª—å–Ω—ã–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º, —ç—Ç–æ –¥–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∫–∞–∂–¥–æ–º —Å—Ç–æ–ª–±—Ü–µ

(df[numeric_columns.columns].isnull().sum()/df[numeric_columns.columns].size).to_frame()
```
